````python
#👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌
# ATTENTION CELLULE CORRECTION A SUPPRIMER

#fonction complétée
def nbAretes(G_mat):
  '''
  Permet d'obtenir le nombre d'arêtes d'un graphe défini par sa matrice d'adjacence
  @param G_mat : matrice d'adjacence d'un graphe G (matrice obligatoirement carrée)
  @return nbAretes : nb d'aretes du graphe
  '''
  ordre = len(G_mat)
  nbAretes = 0
  #parcours de la partie diagonale inférieure de la matrice
  for i in range(ordre):
    for j in range(i+1):
      if G_mat[i][j] != 0:
        nbAretes += 1

  return nbAretes

assert(nbAretes(G0_mat)==13)
````

````python
#👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌👌
# ATTENTION CELLULE CORRECTION A SUPPRIMER

#fonction complétée
def sommetsDegreMax(G_mat):
  '''
  Permet d'obtenir les numero du sommet de plus haut degre d'un graphe défini par sa matrice d'adjacence
  @param G_mat : matrice d'adjacence d'un graphe G (matrice obligatoirement carrée)
  @return numSommetDegreMax : liste des numéros de sommets de degre max
  '''
  ordre = len(G_mat)
  degreMax = -1
  numSommetDegreMax = []
  for i in range(ordre):
    deg_sommet = degre(G_mat,i)
    if deg_sommet >= degreMax :
      if deg_sommet > degreMax :
        numSommetDegreMax = []
      degreMax = deg_sommet
      numSommetDegreMax.append(i)
  return numSommetDegreMax

assert(sommetsDegreMax(G0_mat)==[1,6])
````

## 2. Avec une classe (programmation objet)

On peut également utiliser la **programmation objet** pour implémenter une structure de graphe.

### 2.1 Classe pour les sommets/noeuds

````python
class Node:
  def __init__(self,nom):
    self.nom =nom
    self.traite = False #on utilisera ceci plus tard

  def __repr__(self):
    return f' {self.nom} '
````

### 2.2 Classe pour les arêtes/arcs

````python
class Edge:
  def __init__(self, noeud1, noeud2, poids=''):
    self.extremite1 = noeud1 #objet Node
    self.extremite2 = noeud2 #objet Node
    self.poids = poids

  def __repr__(self):
    return f' [{self.extremite1},{self.extremite2},{self.poids}]\n'
````

### 2.3 Classe pour les graphes

Cette classe utilise les objets noeuds et arêtes précédemment crées.
La classe donnée ci-après permet de créer des arêtes sans avoir crée les sommets auparavant, en les
créant au passage.

````python
class Graphe:
  # graphe non orienté
  def __init__(self, nom):
    self.nom=nom
    self.listeNoeuds=[]
    self.listeAretes=[]

  def __repr__(self):
    #nom du graphe et ordre
    affiche_graphe = self.nom + '\n Ordre : ' + str(len(self.listeNoeuds))+'\n'
    #affichage des sommets
    for noeud in self.listeNoeuds:
      affiche_graphe += repr(noeud)+' '
    affiche_graphe += '\n'
    #affichage des aretes
    affiche_graphe += str(len(self.listeAretes))+' arêtes :\n'
    for arete in self.listeAretes:
      affiche_graphe += "*"+repr(arete)
    return affiche_graphe

  def addNode(self,nomNoeud):
    # vérifie si le noeud du nom entré existe
    for noeud in self.listeNoeuds:
      if nomNoeud == noeud.nom:
        return noeud
    # sinon le crée
    noeud=Node(nomNoeud)
    self.listeNoeuds.append(noeud)
    return noeud

  def addEdge(self,nomNoeud1,nomNoeud2,poids):
    #on cree les noeuds d'extremite si nessecaire
    Noeud1=self.addNode(nomNoeud1)
    Noeud2=self.addNode(nomNoeud2)
    for arete in self.listeAretes:
      # on vérifie si l'arête est déjà là
      if (
          (( arete.extremite1.nom == nomNoeud1 and arete.extremite2.nom == nomNoeud2 )
          or ( arete.extremite1.nom == nomNoeud2 and arete.extremite2.nom == nomNoeud1 ))#aretes non orientees
          and arete.poids == poids
      ):
        return arete
    #sinon on crée une arête entre Noeud1 et Noeud2
    arete=Edge(Noeud1,Noeud2,poids)
    self.listeAretes.append(arete)
    return arete
````

La fonction ci-dessous qui permet d'afficher un graphe donné par sa matrice d'adjacence à l'aide de Graphviz

````python
import graphviz

def afficheGrapheFromMat(M, sommets=[]):
  '''
  Permet d'afficher un graphe défini par sa matrice d'adjacence 
  à l'aide de la librairie Graphviz
  @param M : matrice d'adjacence du graphe
  @param sommets : liste de noms des sommets si différents de 0 à n-1
  @return G graphviz object
  '''

  # la matrice est-elle symetrique ?
  sym = est_symetrique(M)
  if sym: #oui
    G = Graph() #graphe non orienté
  else : #non : graphe orienté
    G = Digraph()
  #print(G) #objet graphviz
  
  ordre = len(M)

  #creation des sommets
  for i in range(ordre):
    if (not(sommets) or (len(sommets)!=ordre)) :#si les noms de sommets ne sont pas définis
      G.node(str(i))#on numérote les sommets
    else : 
      G.node(str(i), label=sommets[i])
  
  #creation des aretes
  aretes = []
  if sym : #graphe non orienté : parcours de la partie inférieure sous la diagonale
    for i in range(ordre):
      for j in range(i+1):
        #print(i,j)
        if M[i][j]!=0:
          aretes.append(str(i)+str(j))
  else : #graphe oriente : parcours complet de la matrice
    for i in range(ordre):
      for j in range(ordre):
        if M[i][j]!=0:
          aretes.append(str(i)+str(j))
  #print("Aretes",aretes)
  G.edges(aretes)

  return G
````

❓ <font color="green" size="3"><b>Q4. a. Ces 1ères fonctions sont données et on ne demande pas nécessairement de comprendre le code, en revanche il est obligatoire de comprendre le fonctionnement en complétant les assertions données.</b></font>

````python
def dictSucc2lstSucc(G_dict_succ, pondere = False):
  """
  Convertit le dictionnaires de listes de successeur d'un graphe
  en liste de listes d'adjacence.
  @param G_dict_succ : dictionnaire de listes de successeurs d'un graphe (orienté ou non)
  @param pondere : Booléen vrai si le graphe est pondéré
  @return G_lst_succ : liste de listes d'adjacence du graphe
  @return sommets : liste des sommets (si non numérotés)
  """
  sommets = list(G_dict_succ.keys())#les clés correspondent aux sommets
  G_lst_succ =[]
  if not(pondere) : #pour les graphes non ponderes
    for s in sommets:#pour chacun des sommets
      lst_voisins_s = []
      voisins = G_dict_succ[s]#on retrouve le nom de ses voisins
      for voisin in voisins:
        lst_voisins_s.append(sommets.index(voisin))#que l'on transforme en indice
      G_lst_succ.append(lst_voisins_s)
  else : #pour les graphes ponderes
    for s in sommets:#pour chacun des sommets
      lst_voisins_ponderations_s = []
      voisins_pond = G_dict_succ[s]#on retrouve les listes voisins/ponderation
      for v_p in voisins_pond:
        lst_voisins_ponderations_s.append((sommets.index(v_p[0]), v_p[1]))#que l'on transforme en indice
      G_lst_succ.append(lst_voisins_ponderations_s)

  return sommets,G_lst_succ



def lstSucc2dictSucc(G_lst_succ, pondere = False, sommets=[]):
    """
    Convertit la liste de listes d'adjacence
    en dictionnaires de listes de successeur.
    @param G_lst_succ  :liste de listes d'adjacence d'un graphe (orienté ou non)
    @param pondere : Booléen vrai si le graphe est pondéré
    @return G_dict_succ : dictionnaire de listes de successeurs du graphe
    """
    G_dict_succ ={}
    ordre = len(G_lst_succ)
    #si la liste des sommets est donnée
    if sommets:
      if not(pondere):
        for i in range(ordre):
          lst_voisins_nommes=[]
          for indice_voisin in G_lst_succ[i]:
            lst_voisins_nommes.append(sommets[indice_voisin])
          G_dict_succ[sommets[i]]=lst_voisins_nommes
      else :
         for i in range(ordre):
          lst_voisins_pondere_nommes=[]
          for indice_voisin_pondere in G_lst_succ[i]:
            lst_voisins_pondere_nommes.append((sommets[indice_voisin_pondere[0]],indice_voisin_pondere[1]))
          G_dict_succ[sommets[i]]=lst_voisins_pondere_nommes      
    else : #on utilise des indices
      for i in range(ordre):
        G_dict_succ[i]=G_lst_succ[i]
    return G_dict_succ
````
